## Testing with the data we could find initially

```python
import json
import codecs

def load_json(file_name):
    with open(file_name, "r", encoding='utf8') as file:
        data = json.load(file)
    return data
```

```python
data1 = load_json("data/comments_topic1.json")
data2 = load_json("data/comments_topic2.json")
data3 = load_json("data/comments_topic3.json")
data4 = load_json("data/comments_topic4.json")

data_list = [data1, data2, data3, data4]

for i in range(len(data_list)):
    print(f"the size of the data{i}:", len(data_list[i]))
```

```python
topic1 = "Miten läheisen kuolemaan liittyvästä asioinnista voisi tehdä helpompaa?"
topic2 = "Mitä mieltä Suomi?"
topic3 = "Mitä mieltä olet Siun soten palveluista?"
topic4 = "Maaseutupoliittisen selonteon Polis-alusta"
```

## Translating the text

```python
from googletrans import Translator

def translate(str):
    translator = Translator()
    trans_text = translator.translate(str, dest='en')
    return trans_text.text
```

```python
topics = [topic1, topic2, topic3, topic4]

translated_topics = [translate(topic) for topic in topics]
translated_topics
```

```python
dfs = []
i = 1
for data, topic, topic_en in zip(data_list, topics, translated_topics):
    df = pd.DataFrame(data)
    df['topic'] = topic
    df['topic_en'] = topic_en
    dfs.append(df)
    i +=1
```

```python
df = pd.concat(dfs, ignore_index=True)
df['is_seed'] = df['is_seed'].astype(int)
df['is_meta'] = df['is_meta'].astype(int)
df
```

```python
# removing the "Taustaväittämä:"
df = df[~df['txt'].str.contains('Taustaväittämä:')]
# df['txt'].str.contains('Taustaväittämä').value_counts()
df.reset_index(drop=True)
```
