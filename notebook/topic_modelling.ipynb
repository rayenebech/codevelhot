## Outputting the coordinates for Streamlit

```python
from typing import List
from umap import UMAP

def export_document_visualization_data(
    topic_model,
    docs: List[str],
    topics: List[int] = None,
    embeddings: np.ndarray = None,
    reduced_embeddings: np.ndarray = None,
    sample: float = None,
) -> dict:
    """Export document visualization data in a D3-friendly JSON format"""
    topic_per_doc = topic_model.topics_

    # Sample the data if specified
    if sample is None or sample > 1:
        sample = 1

    indices = []
    for topic in set(topic_per_doc):
        s = np.where(np.array(topic_per_doc) == topic)[0]
        size = len(s) if len(s) < 100 else int(len(s) * sample)
        indices.extend(np.random.choice(s, size=size, replace=False))
    indices = np.array(indices)

    # Create initial dataframe
    df = pd.DataFrame({
        "topic": [topic_per_doc[index] for index in indices],
        "doc": [docs[index] for index in indices]
    })

    # Handle embeddings
    if reduced_embeddings is None:
        if embeddings is None:
            embeddings_to_reduce = topic_model._extract_embeddings(df.doc.to_list(), method="document")
        else:
            embeddings_to_reduce = embeddings[indices] if sample is not None else embeddings
        
        # Reduce to 2D
        umap_model = UMAP(n_neighbors=10, n_components=2, min_dist=0.0, metric="cosine")
        embeddings_2d = umap_model.fit_transform(embeddings_to_reduce)
    else:
        embeddings_2d = reduced_embeddings[indices] if sample is not None else reduced_embeddings

    # Add coordinates to dataframe
    df["x"] = embeddings_2d[:, 0]
    df["y"] = embeddings_2d[:, 1]

    # Get topic labels/names
    unique_topics = set(topic_per_doc)
    topic_labels = {
        topic: f"Topic {topic}: " + "_".join([word for word, _ in topic_model.get_topic(topic)][:3])
        for topic in unique_topics
    }

    # Create visualization data structure
    viz_data = {
        "points": [
            {
                "x": row.x,
                "y": row.y,
                "topic": int(row.topic),
                "topic_label": topic_labels[row.topic],
                "document": row.doc
            }
            for _, row in df.iterrows()
        ],
        "topics": [
            {
                "topic_id": topic,
                "label": topic_labels[topic],
                "center": {
                    "x": df[df.topic == topic].x.mean(),
                    "y": df[df.topic == topic].y.mean()
                }
            }
            for topic in unique_topics
        ]
    }

    return viz_data
```

```python
viz_data = export_document_visualization_data(topic_model, sentences)
```

```python
import numpy as np

def convert_float32_to_float(data):
    """
    Recursively converts NumPy float32 elements to Python float in a nested dictionary or list.
    
    Args:
        data (dict/list/numpy.ndarray/any): The input data to convert
    
    Returns:
        The converted data with float32 elements converted to Python float
    """
    if isinstance(data, dict):
        # If input is a dictionary, recursively convert each value
        return {key: convert_float32_to_float(value) for key, value in data.items()}
    
    elif isinstance(data, list):
        # If input is a list, recursively convert each element
        return [convert_float32_to_float(item) for item in data]
    
    elif isinstance(data, np.ndarray):
        # If input is a NumPy array, convert float32 to float
        if data.dtype == np.float32:
            return data.astype(float).tolist()
        else:
            return data.tolist()
    
    elif isinstance(data, np.float32):
        # If input is a single float32, convert to Python float
        return float(data)
    
    else:
        # For all other types, return as is
        return data
```
