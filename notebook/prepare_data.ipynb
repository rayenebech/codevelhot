````python
import json

def save_to_json(data, file_path):
    with open(file_path, "w", encoding="utf-8")as fp:
        json.dump(data, fp)

def read_json(file_path):
    with open(file_path, "r", encoding="utf-8")as fp:
        return json.load(fp)
````

## 3rd attempt

Separating inputs based on their topic labels.

````python
df_input = df_labeled[['id', 'Document', 'Name', 'Topic']]
df_input
````

````python
topic_list = read_json('../data/topic_list.json')
topic_list
````

````python
prompt = """
You are an AI facilitator for digital democracy in Finland. Given the comments added by the citizens to the question, 'What do you think, Finland?', identify heavily discussed issues among the comments and actionable recommendations for the policy makers.
The input includes the comments on democracy in Finland as "Document" and a topic label for each comment as "Name", and you are asked to create a dictionary of four issues that summarize the discussion, where keys are the names of the issues identified in the inputs and five values are a list of actionable recommendations for the policy makers that you extract based on the given data.
For each issue, provide 'id's of comments relevant to the issues.
Return your answer as a json format.
"""

model = genai.GenerativeModel("gemini-1.5-flash", generation_config={"response_mime_type": "application/json"})

json_output = {}

for i in range(2):
    df = df_input[df_input['Topic'] == i]
    json_data = df.to_json(orient='records')
    
    topic_name = topic_list[str(i)]
    # print(topic_name)

    payload = {
        "prompt": prompt,
        "data": json_data
    }

    input_text = f"{payload['prompt']}\n\nData:\n{payload['data']}"
    response = model.generate_content(input_text, generation_config=genai.types.GenerationConfig(max_output_tokens=8000))
    json_output[topic_name] = json.loads(response.text)
````

````python
## This is a refined code from the chatGPT.
# Define the model
model = genai.GenerativeModel("gemini-1.5-flash", generation_config={"response_mime_type": "application/json"})

json_output = {}

# Assuming 'df_input' is the DataFrame containing the citizens' comments and 'Topic' column categorizes topics.
# Assuming 'topic_list' is a dictionary mapping topic IDs to topic names.

def process_topic(topic_id):
    try:
        # Filter the data based on the topic
        df = df_input[df_input['Topic'] == topic_id]
        
        # Convert the filtered data to JSON format
        json_data = df.to_json(orient='records')
        
        # Get the corresponding topic name
        topic_name = topic_list.get(str(topic_id), "Unknown Topic")
        
        print(f"Processing topic: {topic_name}")
        
        # Prepare the payload for the model
        payload = {
            "prompt": prompt,
            "data": json_data
        }

        # Format the input text for the model
        input_text = f"{payload['prompt']}\n\nData:\n{payload['data']}"

        # Generate the response
        response = model.generate_content(input_text, generation_config=genai.types.GenerationConfig(max_output_tokens=8000))

        # Parse the response and store it in the output dictionary
        json_output[topic_name] = json.loads(response.text)
        
    except Exception as e:
        print(f"Error processing topic {topic_id}: {e}")

# Process each topic (assuming topics are indexed 0, 1, etc.)
for i in range(28):  # Adjust the range based on the number of topics you have
    process_topic(i)

# Output the final result
print(json_output)
````
