```python
import pandas as pd
import numpy as np
```

```python
import json
import codecs

def load_json(file_name):
    with open(file_name, "r", encoding='utf8') as file:
        data = json.load(file)
    return data
```

```python
data = load_json("comments.json")
print("the size of the data:", len(data))
```

```python
df = pd.DataFrame(data)
df
```
